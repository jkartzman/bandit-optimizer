```python
def f(x):
    return 1+x+x**2
def g(x):
    return 1+x+x**2
def g_prime(x):
    return 1+2*x
```

```python
import numpy as np

def gradient(data,a,b,f,g):
    error_sum_a = 0
    error_sum_b = 0
    for x in data:
        error_sum_a = error_sum_a-2*(f(x)-a*g(b*x))*g(b*x)
        error_sum_b = error_sum_b-2*a*(f(x)-a*g(b*x))*x*g_prime(b*x)
    return error_sum_a/len(data),error_sum_b/len(data)
history_a = [] 
history_b = [] 
a,b = np.random.normal(), np.random.normal()
history_a.append(a)
history_b.append(b)
lr = 0.01
data = np.random.normal(size=1000)
for t in range(1000):
    gradient_a, gradient_b = gradient(data,a,b,f,g)
    a = a-gradient_a*lr
    b = b-gradient_b*lr
    history_a.append(a)
    history_b.append(b)
```
